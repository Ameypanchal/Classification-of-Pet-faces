```python
#import lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as image
import glob
import os
```

```python
images_fp = './images'
```

```python
image_names = [os.path.basename(file) for file in glob.glob(os.path.join(images_fp,'*.jpg'))]
```

```python
image_names
```

```python
labels = [' '.join(name.split('_')[:-1:]) for name in image_names]
```

```python
labels
```

```python
def label_encode(label):
    if label == 'Abyssinian': return 0
    elif label == 'american bulldog': return 1
    elif label == 'american pit bull terrier': return 2
    elif label == 'basset hound': return 3    
    elif label == 'beagle': return 4    
    elif label == 'Birman': return 5    
    elif label == 'Bombay': return 6
    elif label == 'boxer': return 7    
    elif label == 'chihuahua': return 8
    elif label == 'Egyptian Mau': return 9    
    elif label == 'german shorthaired': return 10    
    elif label == 'great pyrenees': return 11   
    elif label == 'British Shorthair': return 12   
    elif label == 'Bengal': return 13   
    elif label == 'english cocker spaniel': return 14   
    elif label == 'english setter': return 15
 
```

```python
import tensorflow as tf
from keras.preprocessing.image import load_img, img_to_array
```

```python
features = []
labels = []
IMG_SIZE = (224, 224)
for name in image_names:
    label =  ' '.join(name.split('_')[:-1:])
    label_encoded = label_encode(label)
    if label_encoded != None:
        img =  load_img(os.path.join(images_fp,name))
        img = tf.image.resize_with_pad(img_to_array(img,dtype = 'uint8'), *IMG_SIZE).numpy().astype('uint8')
        image = np.array(img)
        features.append(image)
        labels.append(label_encoded)
```

```python
features_array = np.array(features)
labels_array = np.array(labels)
```

```python
labels_one_hot = pd.get_dummies(labels_array)
```

```python
labels_one_hot
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
# train -60% + valid-20% test-20%
X_train, X_test, y_train, y_test = train_test_split(features_array, labels_one_hot, test_size = 0.2, random_state= 42)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.25, random_state= 1)
```

```python
from keras import layers, Input, Model
from keras.models import Sequential
from keras.applications import ResNet50
from keras.applications.resnet50 import preprocess_input as ppi
from keras.layers import RandomFlip, RandomRotation, Dense, Dropout
from keras.losses import CategoricalCrossentropy
from keras.optimizers import Adam
```

```python
data_augmentation = Sequential([RandomFlip('horizontal_and_vertical'), RandomRotation(0.2)])
prediction_layers = Dense(16, activation = 'softmax')
```

```python
resnet_model = ResNet50(include_top = False, pooling='avg', weights='imagenet')
resnet_model.trainable=False
preprocess_input = ppi
```

```python
inputs =  Input(shape=(224,224,3))
x = data_augmentation(inputs)
x = preprocess_input(x)
x = resnet_model(x, training=False)
x = Dropout(0.2)(x)
outputs = prediction_layers(x)
model = Model(inputs, outputs)
```

```python
model.compile(optimizer=Adam(), loss=CategoricalCrossentropy(), metrics=['accuracy'])
model.summary()
```

```python
model_history = model.fit(x=X_train, y=y_train, validation_data=(X_val,y_val), epochs=5)
```

```python
acc = model_history.history['accuracy']
val_acc = model_history.history['val_accuracy']
loss = model_history.history['loss']
val_loss = model_history.history['val_loss']
```

```python
epochs_range = range(5)
plt.figure(figsize=(15,8))

plt.subplot(1,2,1)
plt.plot(epochs_range, acc, label='Training accuracy')
plt.plot(epochs_range, val_acc, label='Validation accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1,2,2)
plt.plot(epochs_range, loss, label='Training loss')
plt.plot(epochs_range, val_loss, label='Validation loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
```

```python
model.evaluate(X_test, y_test)
```
